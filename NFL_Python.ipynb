```python
nfl_filter = df[['posteam', 'play_type', 'quarter_seconds_remaining', 'pass_length', 'pass_location', 'run_location', 'ydstogo', 'first_down_rush',
                 'yrdln', 'game_id', 'down', 'score_differential_post', 'incomplete_pass', 'complete_pass', 'rush_attempt', 'pass_attempt', 
                 'defteam', 'qtr', 'game_seconds_remaining', 'posteam_timeouts_remaining', 'defteam_timeouts_remaining',
                 'epa', 'wpa', 'yardline_100', 'half_seconds_remaining', 'goal_to_go', 'side_of_field',
                 'shotgun', 'drive', 'no_huddle', 'penalty']]
```

```python
df[((df['down'] == 1) & (df['ydstogo'] <= 12))]
```

```python
mask1 = df['down'] == 1
mask11 = df['ydstogo'] <= 12
mask2 = df['down'] == 2 
mask22 = df['ydstogo'] <= 5
mask3 = df['down'] == 3 
mask33 = df['ydstogo'] <= 2 
mask4 = df['down'] == 4 
mask44 = df['ydstogo'] <= 1
```

```python
df['down_ydstogo'] = 0
df.loc[((mask1 & mask11) | (mask2 & mask22) | (mask3 & mask33) | (mask4 & mask44)), 'down_ydstogo'] = 1
```

```python
df[df['down_ydstogo'] == 1]
```

```python
df['time_score'] = df['game_seconds_remaining']
df.loc[df['game_seconds_remaining'] < 900, 'time_score'] = \
    df['score_differential_post'] * df['score_differential_post'] * df['game_seconds_remaining']
```

```python
df = df.loc[(df['play_type'] == 'pass') | (df['play_type'] == 'run')]
```

```python
df = df.loc[df['ydstogo'] < 36]
```

```python
df = df[~df[['pass_location', 'run_location']].notnull().all(1)]
```

```python
df = df[~((df['incomplete_pass'] == 1) & (df['complete_pass'] == 1) & (df['play_type'] == 'pass'))]
```

```python
df['pass_length'] = df['pass_length'].fillna('n/a')
df['pass_location'] = df['pass_location'].fillna('n/a')
df['run_location'] = df['run_location'].fillna('n/a')
df['defteam'] = df['defteam'].fillna('n/a')
df['side_of_field'] = df['side_of_field'].fillna('n/a')
df['yrdln'] = df['yrdln'].fillna('n/a')
```

```python
df
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df
```

```python
df['posteam_f'] = pd.factorize(df['posteam'])[0]
df
```

```python
df['play_type_f'] = pd.factorize(df['play_type'])[0]
df['pass_length_f'] = pd.factorize(df['pass_length'])[0]
df['pass_location_f'] = pd.factorize(df['pass_location'])[0]
df['run_location_f'] = pd.factorize(df['run_location'])[0]
df['defteam_f'] = pd.factorize(df['defteam'])[0]
df['side_of_field_f'] = pd.factorize(df['side_of_field'])[0]
df['yrdln_f'] = pd.factorize(df['yrdln'])[0]
```

```python
rush_pass_ratio = df.groupby('posteam').sum()['play_type_f'] / df.groupby('posteam').count()['play_type_f']
```

```python
rp_ratio_dict = rush_pass_ratio.to_dict()
```

```python
df['posteam_rp_ratio'] = df['posteam'].map(rp_ratio_dict)
```
